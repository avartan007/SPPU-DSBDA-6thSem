```python
import pandas as pd

# Load the dataset
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv'
columns = ['sepal-length','sepal-width','petal-length','petal-width','class']
df = pd.read_csv(url, names=columns)

# Display info
print(df.shape)
print(df.head())
print(df.info())

# Summary statistics
print(df.describe())
```

```python
# Group by 'class' and get summary statistics
grouped_stats = df.groupby('class').agg(['mean', 'median', 'min', 'max', 'std'])
print("\nGrouped Summary Statistics by Class:")
print(grouped_stats)
```

```python
#Creating list of sepal-length values grouped by class
species_groups = df.groupby('class')['sepal-length'].apply(list)
print("\nList of 'sepal-length' values for each species:")
print(species_groups.to_dict())
```

```python
setosa = df[df['class'] == 'Iris-setosa'].describe()
versicolor = df[df['class'] == 'Iris-versicolor'].describe()
virginica = df[df['class'] == 'Iris-virginica'].describe()
```

```python
print("\nDescriptive Stats for Iris-setosa:")
display(setosa)

print("\nDescriptive Stats for Iris-versicolor:")
display(versicolor)

print("\nDescriptive Stats for Iris-virginica:")
display(virginica)
```
